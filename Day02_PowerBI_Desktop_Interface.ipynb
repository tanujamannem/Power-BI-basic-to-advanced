# ‚≠ê 1. Overview of Power BI Desktop Interface

Power BI Desktop is the **primary tool for building reports and dashboards**. It consists of:

- Ribbon
- Report view
- Data view
- Model view
- Fields pane
- Visualizations pane

You can **connect, transform, model, and visualize data** all in one place.

---

# ‚≠ê 2. Ribbon Overview

The Ribbon contains the main menu options:

- **Home:** Import data, transform data, manage relationships  
- **Insert:** Add visuals, shapes, text boxes  
- **Modeling:** Create measures, calculated columns, manage roles  
- **View:** Themes, page layout, performance analyzer  
- **Help:** Documentation, tutorials, updates

---

# ‚≠ê 3. Report, Data & Model Views

- **Report View:** Build and customize visuals, add slicers, charts, tables, maps  
- **Data View:** Explore tables and columns, check data types, calculated columns  
- **Model View:** Visualize relationships between tables, manage primary/foreign keys and relationship cardinality  

---

# ‚≠ê 4. Fields Pane

The Fields pane shows:

- All tables imported  
- Columns inside each table  
- Measures and calculated columns  

You can **drag fields directly to report canvas** to create visuals.

---

# ‚≠ê 5. Visualizations Pane

The Visualizations pane allows you to:

- Select chart types (bar, line, pie, scatter, map, table, matrix)  
- Configure fields for visuals (Values, Axis, Legend, Tooltips)  
- Apply formatting: colors, labels, data categories  
- Add slicers and filters

---

# ‚≠ê 6. Themes and Formatting Basics

- **Themes:** Apply prebuilt or custom color schemes  
- **Formatting options:** Font size, color, border, background  
- **Consistency:** Helps maintain professional dashboards  

---

# ‚≠ê 7. Example Scenario

A sales analyst wants to visualize **monthly sales by product category**:

1. Open **Power BI Desktop**  
2. Load the sales dataset  
3. Drag **Product Category** to the axis and **Sales Amount** to values  
4. Choose a **column chart**  
5. Apply formatting: change colors, add data labels  
6. Add a **slicer** for region to filter the data interactively  

---

# ‚≠ê 8. Key Takeaways

- Ribbon provides all main functionalities  
- Three views (Report, Data, Model) are essential for development  
- Fields and Visualizations panes are used to drag/drop fields into visuals  
- Formatting and themes improve dashboard readability  

---

# ‚≠ê 9. Interview Questions

**Q1: What are the three main views in Power BI Desktop?**  
Report, Data, and Model views.

**Q2: What is the purpose of the Fields pane?**  
To show tables, columns, and measures, and allow dragging fields into visuals.

**Q3: Name some main sections of the Ribbon.**  
Home, Insert, Modeling, View, Help.

**Q4: What is the use of the Visualizations pane?**  
To select chart types, configure fields, and apply formatting.

**Q5: How do you change the theme of a report?**  
Go to View ‚Üí Themes ‚Üí select a prebuilt or custom theme.

**Q6: What is the purpose of Model view?**  
To manage relationships, keys, and table structures.

**Q7: Can you add slicers in Power BI Desktop?**  
Yes, slicers allow interactive filtering of visuals.

**Q8: What is the difference between Report and Data view?**  
Report view is for creating visuals; Data view is for inspecting tables and columns.

**Q9: Why is formatting important in Power BI reports?**  
Improves readability, consistency, and professional appearance.

**Q10: How can a user explore relationships between tables?**  
Using Model view to visualize and manage relationships.
