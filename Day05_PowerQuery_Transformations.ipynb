# ‚≠ê 1. Overview of Transformations

Transformations in Power Query allow you to **reshape, clean, and prepare data** for analysis.  
Common transformations include **splitting, merging, appending tables, and creating conditional columns**.

---

# ‚≠ê 2. Split Columns

- Split columns by delimiter (comma, space, hyphen)  
- Split columns by number of characters  
- Useful for extracting meaningful information from combined fields (e.g., "City, State")  

---

# ‚≠ê 3. Merge Tables

- Combine two tables **horizontally** based on a common column  
- Similar to SQL JOIN operations  
- Types of joins: Inner, Left Outer, Right Outer, Full Outer, Anti, etc.  
- Example: Merge Sales table with Product table to add product details  

---

# ‚≠ê 4. Append Tables

- Combine two or more tables **vertically** (stack rows)  
- Column names and data types should match  
- Useful for combining monthly or yearly data into a single table  

---

# ‚≠ê 5. Conditional Columns

- Create new columns based on conditions (if/then logic)  
- Example: Create a "Sales Category" column:
  - If Sales > 1000 ‚Üí "High"  
  - Else ‚Üí "Low"  
- Found under **Add Column ‚Üí Conditional Column**

---

# ‚≠ê 6. Text Transformations

- Change case: Upper, Lower, Capitalize Each Word  
- Trim spaces, clean non-printable characters  
- Extract text: Left, Right, Mid, or by delimiter  

---

# ‚≠ê 7. Date Transformations

- Extract parts: Year, Month, Day, Quarter, Weekday  
- Add or subtract days, months, or years  
- Useful for **time-based analysis and creating date hierarchies**

---

# ‚≠ê 8. Example Scenario

A sales dataset requires multiple transformations:

1. Split **Customer Name** into First and Last Name  
2. Merge Sales table with Product table to get category details  
3. Append data from January and February into one table  
4. Create a **Sales Category** column based on Sales Amount  
5. Extract **Year** and **Month** from Order Date for analysis  

---

# ‚≠ê 9. Key Takeaways

- Transformations help **clean, reshape, and enrich data**  
- Splitting, merging, and appending tables are essential for combining datasets  
- Conditional columns allow **dynamic categorization**  
- Text and date transformations improve **reporting and analysis accuracy**

---

# ‚≠ê 10. Interview Questions

**Q1: What is the purpose of transformations in Power Query?**  
To clean, reshape, and prepare data for analysis.

**Q2: How do you split a column by delimiter?**  
Select the column ‚Üí Home or Transform ‚Üí Split Column ‚Üí By Delimiter.

**Q3: What is the difference between Merge and Append?**  
Merge combines tables horizontally (like JOIN), Append combines tables vertically (stack rows).

**Q4: Name some types of joins in Merge.**  
Inner, Left Outer, Right Outer, Full Outer, Anti.

**Q5: What is a conditional column?**  
A column created using if/then logic based on conditions.

**Q6: How can you transform text in Power Query?**  
Change case, trim spaces, clean characters, extract text.

**Q7: How do you extract the year or month from a date column?**  
Select the date column ‚Üí Transform ‚Üí Date ‚Üí Year or Month.

**Q8: Why is it useful to append tables?**  
To combine multiple datasets with the same structure into one table.

**Q9: Can Power Query transformations be undone or edited?**  
Yes, each step is recorded in Applied Steps and can be edited or deleted.

**Q10: Why are transformations important before loading data into Power BI?**  
They ensure clean, accurate, and structured data for modeling and visualization.
