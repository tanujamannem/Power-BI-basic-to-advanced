# ‚≠ê 1. Overview of Data Profiling

Data profiling is the process of **analyzing the quality, structure, and content of your data**.  
It helps identify **errors, inconsistencies, and missing values** before modeling.

---

# ‚≠ê 2. Enabling Data Profiling in Power Query

1. Open **Power Query Editor**  
2. Go to **View ‚Üí Column Quality, Column Distribution, Column Profile**  
3. These features provide:
   - **Column Quality:** Percentage of valid, error, and empty values  
   - **Column Distribution:** Frequency of distinct values  
   - **Column Profile:** Detailed statistics for numeric and text columns  

---

# ‚≠ê 3. Column Statistics

For numeric columns, Power Query provides:  
- Minimum, Maximum, Average  
- Standard deviation  
- Count of distinct values  
- Count of errors and empty values  

For text columns:  
- Count of distinct and unique values  
- Count of errors and blanks  

---

# ‚≠ê 4. Error Detection

- Rows with errors are highlighted  
- Common issues include:  
  - Data type mismatches  
  - Missing or null values  
  - Unexpected characters in numeric fields  
- Errors can be **removed, replaced, or corrected**  

---

# ‚≠ê 5. Data Cleaning Tips

- Handle missing values: Replace with default values or remove rows  
- Standardize text: Trim spaces, fix case, remove special characters  
- Correct data types before loading  
- Remove duplicates and irrelevant columns  

---

# ‚≠ê 6. Example Scenario

A sales dataset has inconsistencies:

1. Open Power Query Editor ‚Üí Sales table  
2. Enable **Column Quality** and **Column Profile**  
3. Identify:
   - Rows with missing Region  
   - Numeric columns with errors  
   - Duplicate Customer IDs  
4. Clean data:
   - Replace missing Region with "Unknown"  
   - Correct data type for Sales Amount  
   - Remove duplicate Customer IDs  
5. Close & Apply to load clean data  

---

# ‚≠ê 7. Key Takeaways

- Data profiling ensures **data accuracy and reliability**  
- Column statistics help detect **outliers and inconsistencies**  
- Errors, missing values, and duplicates should be addressed before modeling  
- Clean data leads to **better visuals and correct calculations**

# ‚≠ê 8. Interview Questions

**Q1: What is data profiling in Power BI?**  
The process of analyzing data quality, structure, and content before modeling.

**Q2: How do you enable column profiling in Power Query?**  
Go to View ‚Üí Column Quality, Column Distribution, Column Profile.

**Q3: What information does Column Quality provide?**  
Percentage of valid, error, and empty values in a column.

**Q4: What can Column Distribution show for text columns?**  
Frequency of distinct values and unique value count.

**Q5: How can errors in data be handled?**  
Remove, replace, or correct them in Power Query Editor.

**Q6: Why is it important to check data types during profiling?**  
Incorrect data types can cause calculation and visualization errors.

**Q7: What steps can you take for missing values?**  
Replace with default values or remove affected rows.

**Q8: Why remove duplicates during data profiling?**  
To ensure accurate aggregation and prevent skewed analysis.

**Q9: How does data profiling improve report quality?**  
It ensures clean and consistent data, resulting in accurate visuals and calculations.

**Q10: Can Power Query highlight errors automatically?**  
Yes, errors are highlighted in Column Quality and Column Profile for easy identification.
