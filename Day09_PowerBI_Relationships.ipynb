# ‚≠ê 1. Overview of Relationships

Relationships in Power BI define **how tables are connected** so you can perform accurate calculations and create meaningful reports.  
They allow **filtering and aggregation** across related tables.

---

# ‚≠ê 2. Types of Relationships

1. **One-to-Many (1:*):**  
   - A single row in one table matches multiple rows in another.  
   - Example: Product table ‚Üí Sales table.

2. **Many-to-One (*:1):**  
   - Reverse of One-to-Many.  
   - Example: Sales table ‚Üí Product table (same as above, depending on direction).

3. **Many-to-Many (*:*):**  
   - Multiple rows in both tables match each other.  
   - Less common; use with caution.  
   - Example: Students ‚Üî Courses (many students enroll in many courses).

---

# ‚≠ê 3. Active vs Inactive Relationships

- **Active Relationship:** Used by default in calculations.  
- **Inactive Relationship:** Exists but not used unless explicitly activated using DAX (`USERELATIONSHIP`).  
- **Example:** If a Sales table has multiple date columns (OrderDate, DeliveryDate), only one can be active at a time.

---

# ‚≠ê 4. Role-Playing Dimensions

- A single dimension table can be used in multiple roles.  
- Example: Date table for Order Date and Delivery Date in Sales table.  
- Activate one relationship at a time; use DAX `USERELATIONSHIP` for inactive relationships.

---

# ‚≠ê 5. Creating Relationships in Power BI (Step-by-Step)

1. Go to **Model view** in Power BI Desktop.  
2. Identify **primary key** and **foreign key** columns.  
3. Drag primary key ‚Üí Drop on foreign key column in another table.  
4. Relationship settings window appears:  
   - **Cardinality:** One-to-many, Many-to-one, Many-to-many  
   - **Cross filter direction:** Single or Both  
   - **Make this relationship active** checkbox  
5. Click **OK** to create the relationship.  
6. Repeat for all necessary tables.

---

# ‚≠ê 6. Example Scenario

Company dataset: Sales table, Product table, Customer table, Date table.

1. Go to **Model view**  
2. Drag **ProductID** from Product ‚Üí Sales table ProductID  
   - Cardinality: One-to-Many  
   - Cross filter: Single  
   - Active: Checked  
3. Drag **CustomerID** from Customer ‚Üí Sales table CustomerID  
4. Drag **Date** from Date table ‚Üí Sales table OrderDate  
5. Document relationships: ensure each table is correctly connected with proper cardinality and direction.

---

# ‚≠ê 7. Best Practices

- Prefer **One-to-Many relationships** wherever possible.  
- Use **active relationships** for default calculations.  
- Inactive relationships can be activated via **DAX USERELATIONSHIP**.  
- Avoid unnecessary many-to-many relationships unless required.  
- Maintain clear **primary and foreign keys** for each table.

---

# ‚≠ê 8. Key Takeaways

- Relationships enable **filtering, aggregation, and correct calculations**.  
- Correct cardinality and cross-filter direction are crucial.  
- Role-playing dimensions allow **flexible use of tables** without creating duplicates.  
- Always document relationships for transparency and troubleshooting.

# ‚≠ê 9. Interview Questions

**Q1: What is a relationship in Power BI?**  
A connection between tables that allows filtering and aggregation across them.

**Q2: What is a one-to-many relationship?**  
A single row in one table corresponds to multiple rows in another table.

**Q3: What is a many-to-many relationship?**  
Multiple rows in both tables can match each other.

**Q4: What is the difference between active and inactive relationships?**  
Active is used by default; inactive exists but must be activated using DAX.

**Q5: How do you create a relationship in Power BI?**  
Model view ‚Üí Drag primary key ‚Üí Drop on foreign key ‚Üí Set cardinality and cross-filter ‚Üí OK.

**Q6: What is a role-playing dimension?**  
A table used in multiple contexts, e.g., Date table used for Order Date and Delivery Date.

**Q7: How do you use an inactive relationship in calculations?**  
Activate it using the DAX function `USERELATIONSHIP`.

**Q8: What is cross filter direction?**  
Determines how filters propagate between related tables: Single or Both directions.

**Q9: Why should many-to-many relationships be used cautiously?**  
They can create ambiguity and slow performance in calculations.

**Q10: What are best practices for managing relationships?**  
Use clear keys, prefer one-to-many, document relationships, and activate only required relationships.
