# ‚≠ê 1. Overview of Power Query Editor

Power Query Editor is a **data transformation and preparation tool** in Power BI.  
It allows you to **clean, reshape, and manipulate data** before loading it into the model.

---

# ‚≠ê 2. Opening Power Query Editor

To open Power Query Editor:  
1. Go to **Home ‚Üí Transform Data** in Power BI Desktop  
2. The Power Query Editor window will open, showing **Applied Steps**, preview of your tables, and query pane

---

# ‚≠ê 3. Interface Components

- **Queries Pane:** List of all imported tables/queries  
- **Data Preview:** Shows a snapshot of the table  
- **Ribbon:** Options for transforming data (Remove Rows, Split Columns, Replace Values, etc.)  
- **Applied Steps Pane:** Tracks every transformation step, which can be edited or deleted

---

# ‚≠ê 4. Common Data Cleaning Operations

- **Remove duplicates** ‚Äì Eliminate repeated rows  
- **Remove errors** ‚Äì Detect and remove rows with errors  
- **Rename columns** ‚Äì For clarity and consistency  
- **Change data types** ‚Äì Ensure correct numeric, text, or date types  
- **Replace values** ‚Äì Standardize inconsistent data  

---

# ‚≠ê 5. Filters in Power Query

- Filter rows based on conditions (e.g., Sales > 1000)  
- Text filters (contains, starts with, ends with)  
- Date filters (before, after, between)  
- Numeric filters (greater than, less than, equal to)  

---

# ‚≠ê 6. Applied Steps Concept

- Every transformation creates a **step in Applied Steps**  
- Steps are **reversible** ‚Äì you can edit, delete, or reorder them  
- Helps maintain a **clean and reproducible workflow**

---

# ‚≠ê 7. Example Scenario

A dataset contains sales data with some errors and duplicates:

1. Open **Power Query Editor**  
2. Select the Sales table  
3. Remove duplicate rows  
4. Replace missing Region values with "Unknown"  
5. Change **Sales Amount** column to Decimal Number  
6. Filter out rows where Quantity ‚â§ 0  
7. Rename columns for clarity  
8. Close & Apply to load clean data into Power BI

---

# ‚≠ê 8. Key Takeaways

- Power Query Editor is essential for **data cleaning and transformation**  
- Applied Steps track all transformations for easy edits  
- Filtering, removing duplicates, and replacing values ensures **data quality**  
- Always clean and prepare data before modeling for accurate analysis

# ‚≠ê 9. Interview Questions

**Q1: What is Power Query Editor used for?**  
For cleaning, transforming, and preparing data before loading it into the model.

**Q2: How do you open Power Query Editor in Power BI?**  
Home ‚Üí Transform Data.

**Q3: What is the Applied Steps pane?**  
It tracks all transformations performed on a query in a step-by-step manner.

**Q4: How can you remove duplicate rows in Power Query Editor?**  
Select the table ‚Üí Home ‚Üí Remove Rows ‚Üí Remove Duplicates.

**Q5: How do you replace values in a column?**  
Right-click the column ‚Üí Replace Values ‚Üí Enter old and new value.

**Q6: Why is it important to change data types in Power Query?**  
Correct data types are essential for accurate calculations and visuals.

**Q7: Can Applied Steps be edited or deleted?**  
Yes, each step can be modified, reordered, or removed.

**Q8: What types of filters are available in Power Query?**  
Text filters, numeric filters, and date filters.

**Q9: Why is data cleaning important before loading into Power BI?**  
Ensures accurate analysis, reduces errors, and improves model quality.

**Q10: What happens after you click Close & Apply in Power Query Editor?**  
Cleaned and transformed data is loaded into Power BI for modeling and visualization.
