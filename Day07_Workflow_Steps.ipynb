# ‚≠ê 2. Step-by-Step Workflow

1. **Select a dataset**  
   - Choose a sample dataset (Excel or CSV) with numeric, text, and date columns.  
   - Example: Sales dataset with columns: SalesID, Product, Quantity, Price, OrderDate, Region.

2. **Understand the dataset**  
   - Open Power BI Desktop ‚Üí Load the dataset  
   - Go to **Data view** ‚Üí Check column names and data types  
   - Identify numeric, text, date, and categorical columns  
   - Note missing values, duplicates, or inconsistent data  

3. **Handle Missing Values**  
   - Open **Transform Data ‚Üí Power Query Editor**  
   - For numeric columns: Select column ‚Üí **Transform ‚Üí Replace Values ‚Üí Replace null with 0 or average**  
   - For text columns: Select column ‚Üí **Transform ‚Üí Replace Values ‚Üí Replace null with "Unknown"**  
   - For date columns: Select column ‚Üí **Transform ‚Üí Replace Values ‚Üí Replace null with default date**  

4. **Remove Duplicates (explicit steps)**  
   - Open **Transform Data ‚Üí Power Query Editor**  
   - Select the column(s) to check for duplicates (e.g., SalesID)  
   - Go to **Home tab ‚Üí Remove Rows ‚Üí Remove Duplicates**  
   - Power BI will keep the first occurrence of each unique value and remove all duplicates  
   - Apply the same for multiple columns if needed  

5. **Rename Columns**  
   - In Power Query Editor: Right-click column ‚Üí **Rename** ‚Üí Enter a clear, meaningful name  
   - Example: "Prod_Name" ‚Üí "Product Name", "Qty" ‚Üí "Quantity"

6. **Change Data Types**  
   - Select column ‚Üí **Transform ‚Üí Data Type ‚Üí Choose appropriate type (Text, Decimal Number, Date/Time, etc.)**

7. **Split Columns (if required)**  
   - Select column ‚Üí **Transform ‚Üí Split Column ‚Üí By Delimiter**  
   - Example: Split "City, State" into two separate columns

8. **Merge or Append Tables**  
   - **Merge:** Home ‚Üí Combine ‚Üí Merge Queries ‚Üí Select tables and join key ‚Üí Choose join type  
   - **Append:** Home ‚Üí Combine ‚Üí Append Queries ‚Üí Select tables to stack vertically  

9. **Data Profiling / Quality Check**  
   - Go to **View ‚Üí Column Quality, Column Distribution, Column Profile**  
   - Identify errors, blanks, or outliers in numeric and text columns  
   - Document which issues were found and how you fixed them  

10. **Document Workflow**  
    - Step 1: Import dataset  
    - Step 2: Handle missing values  
    - Step 3: Remove duplicates using **Home ‚Üí Remove Rows ‚Üí Remove Duplicates**  
    - Step 4: Rename columns  
    - Step 5: Change data types  
    - Step 6: Split/Merge/Append tables if necessary  
    - Step 7: Data profiling for errors and outliers  
    - Step 8: Dataset ready for modeling

# ‚≠ê 4. Interview Questions

**Q1: How do you remove duplicates in Power BI Desktop?**  
Select the column(s) ‚Üí **Home ‚Üí Remove Rows ‚Üí Remove Duplicates**. Power BI keeps the first occurrence and deletes duplicates.

**Q2: How do you handle missing numeric values?**  
Transform column ‚Üí Replace null with 0 or column average.

**Q3: How do you handle missing text values?**  
Transform column ‚Üí Replace null with "Unknown".

**Q4: Why rename columns clearly?**  
Improves readability, understanding, and consistency in reports.

**Q5: How do you ensure correct data types?**  
Transform column ‚Üí Data Type ‚Üí Select appropriate type (Text, Number, Date/Time).

**Q6: When should you split columns?**  
When a single column contains multiple pieces of information, e.g., "City, State".

**Q7: When should you merge tables?**  
To combine tables horizontally based on a common key.

**Q8: When should you append tables?**  
To combine tables vertically (stack rows) when structure is similar.

**Q9: What is the purpose of data profiling?**  
To detect errors, blanks, duplicates, outliers, and ensure data quality before modeling.

**Q10: Why is documenting the workflow important?**  
For reproducibility, knowledge transfer, and explaining steps in interviews.
