# â­ 1. Overview
DAX (Data Analysis Expressions) is the formula language used in Power BI to create calculations and logic. Todayâ€™s focus is on basic aggregation and logical functions.

# â­ 2. Aggregation Functions
- SUM: Adds all values in a numeric column  
  Syntax: SUM(Table[Column])  
  Example: Total Quantity := SUM(Sales[Quantity])

- AVERAGE: Returns the average of numeric values  
  Syntax: AVERAGE(Table[Column])  
  Example: Avg Price := AVERAGE(Sales[Price])

- COUNT: Counts all rows with a value  
  Syntax: COUNT(Table[Column])  
  Example: Order Count := COUNT(Sales[SalesID])

- DISTINCTCOUNT: Counts unique values  
  Syntax: DISTINCTCOUNT(Table[Column])  
  Example: Unique Products := DISTINCTCOUNT(Sales[ProductID])

# â­ 3. Logical Functions
- IF: Conditional logic  
  Syntax: IF(condition, result_if_true, result_if_false)  
  Example: High Sales := IF(Sales[TotalPrice] > 500, "Yes", "No")

- SWITCH: Multiple conditions  
  Syntax: SWITCH(expression, value1, result1, value2, result2, ..., else_result)  
  Example:  
  Sales Category := SWITCH(TRUE(),  
                           Sales[TotalPrice] > 1000, "High",  
                           Sales[TotalPrice] > 500, "Medium",  
                           "Low")

- ISBLANK: Checks for empty values  
  Syntax: ISBLANK(Table[Column])  
  Example: Missing Quantity := IF(ISBLANK(Sales[Quantity]), "Yes", "No")

# â­ 4. Combining Aggregation and Logic
You can combine SUM, AVERAGE, COUNT with IF for conditional calculations.  
Example: Total Sales for orders > 500  
Large Orders := CALCULATE(SUM(Sales[TotalPrice]), Sales[TotalPrice] > 500)  

Example: Count orders with quantity > 10  
Orders >10 := COUNTROWS(FILTER(Sales, Sales[Quantity] > 10))

# â­ 6. Key Takeaways
- Aggregation functions summarize numeric data (SUM, AVERAGE, COUNT, DISTINCTCOUNT)  
- Logical functions allow conditional calculations (IF, SWITCH, ISBLANK)  
- Combining aggregation and logic enables powerful analysis for reports  
- Understanding row context vs filter context is important for accurate results

# â­ 7. Interview Questions
**Q1: What is DAX in Power BI?**  
A formula language used for creating calculations and logic in Power BI.

**Q2: What is the difference between SUM and SUMX?**  
SUM adds values in a column directly; SUMX evaluates an expression row by row and then sums.

**Q3: How does the IF function work?**  
It evaluates a condition and returns a value if true or false.

**Q4: What is the SWITCH function used for?**  
To handle multiple conditions or categories in a single calculation.

**Q5: What is ISBLANK used for?**  
To check if a column value is blank or empty.

**Q6: How can you combine aggregation and logic?**  
Use functions like SUM or COUNT with IF or FILTER to calculate conditional aggregates.

**Q7: What is DISTINCTCOUNT?**  
Counts the number of unique values in a column.

**Q8: How do you calculate conditional sums in DAX?**  
Use CALCULATE with a filter condition, e.g., CALCULATE(SUM(Sales[TotalPrice]), Sales[TotalPrice] > 500).

**Q9: What is the difference between row context and filter context?**  
Row context applies to individual rows (calculated columns); filter context applies to aggregations in measures or visuals.

**Q10: Can logical functions be used in measures?**  
Yes, logical functions like IF, SWITCH, and ISBLANK can be used in measures and calculated columns.
