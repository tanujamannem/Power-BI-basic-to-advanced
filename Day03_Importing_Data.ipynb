# ‚≠ê 1. Overview of Data Import in Power BI

Importing data is the first step in creating reports and dashboards.  
Power BI allows you to **connect to multiple data sources** and prepare data for analysis.

---

# ‚≠ê 2. Common Data Sources

Power BI supports a variety of sources, including:

- **Excel** ‚Äì .xlsx or .xls files  
- **CSV/Text files** ‚Äì Comma or tab-separated  
- **SQL Server** ‚Äì On-premises or cloud databases  
- **Web** ‚Äì Import data from web URLs or APIs  
- **SharePoint** ‚Äì Lists and document libraries  
- **Other sources** ‚Äì Azure, Google Analytics, Salesforce, OData feeds

---

# ‚≠ê 3. Data Types and Conversions

After importing data, check **data types** in the Fields pane:

- **Numeric:** Whole Number, Decimal Number  
- **Text:** String or categorical values  
- **Date/Time:** Date, DateTime, Time  
- **Boolean:** True/False values  

You can **change data types** in the Data view or Power Query Editor.

---

# ‚≠ê 4. Load vs Transform

- **Load:** Directly imports the data as-is into Power BI  
- **Transform (Power Query Editor):** Clean, reshape, and prepare data before loading  

Typical transformations include:  
- Remove duplicates  
- Rename columns  
- Change data types  
- Split or merge columns  

---

# ‚≠ê 5. Example Scenario

A company wants to analyze **sales data from Excel and SQL Server**:

1. Open Power BI Desktop  
2. Go to **Home ‚Üí Get Data ‚Üí Excel** and load the file  
3. Go to **Home ‚Üí Get Data ‚Üí SQL Server** and connect using credentials  
4. Preview tables, select necessary columns  
5. Use **Transform Data** to clean missing or incorrect values  
6. Load data into the model  

---

# ‚≠ê 6. Key Takeaways

- Power BI connects to multiple data sources easily  
- Always check data types and convert if needed  
- Use **Transform Data** for cleaning before loading to the model  
- Properly prepared data ensures smooth modeling and visualization  

---

# ‚≠ê 7. Interview Questions

**Q1: What is the difference between Load and Transform in Power BI?**  
Load imports data directly; Transform allows cleaning and shaping data before import.

**Q2: Name some common data sources in Power BI.**  
Excel, CSV, SQL Server, Web, SharePoint, Azure, Salesforce.

**Q3: How can you change the data type of a column?**  
In Data view or Power Query Editor, select the column and change its data type.

**Q4: What data types are supported in Power BI?**  
Numeric, Text, Date/Time, Boolean.

**Q5: Can Power BI connect to web data sources?**  
Yes, through web URLs or APIs.

**Q6: What is Power Query Editor used for?**  
For cleaning, transforming, and shaping data before loading it into the model.

**Q7: How do you remove duplicate rows in Power BI?**  
Use Power Query Editor ‚Üí Remove Duplicates option.

**Q8: Why is it important to check data types after importing data?**  
Incorrect data types can cause errors in calculations and visuals.

**Q9: Can Power BI connect to SQL Server databases?**  
Yes, both on-premises and cloud SQL Server instances.

**Q10: What is the advantage of transforming data before loading?**  
It ensures clean, accurate, and properly structured data for modeling and visualization.
